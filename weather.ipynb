## Visualization

```python
px.violin(df, x='Weather', title='Weather')
```

```python
plt.hist(df['Visibility_km'],color='red')
```

```python
px.box(df, x='Wind Speed_km/h', title='Wind Speed')
```

```python
px.treemap(df, path=['Weather', 'Wind Speed_km/h'],values='Visibility_km', title='Weather and Wind Speed')
```

```python
sns.jointplot(df, x='Visibility_km', y='Temp_C', kind='hex')
```

```python
sns.pairplot(df)
```

```python
px.histogram(df, x='Dew Point Temp_C', y='Temp_C', color='Weather', title='Dew Point Temp and Temp')
```

```python
px.bar(df,x='Wind Speed_km/h', y='Visibility_km',color='Weather',title='Wind Speed and Visibility')
```

```python
px.line(df, x='Weather', y='Wind Speed_km/h', color= 'Weather',title='Wind speed acc to Weather',height=800,width=1000)
```

```python
px.scatter_matrix(df, dimensions=['Wind Speed_km/h', 'Visibility_km', 'Temp_C'], title='Scatter Matrix')
```
